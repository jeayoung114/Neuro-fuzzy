```python
import numpy as np
import pandas as pd
import math
from math import exp
from sklearn.preprocessing import normalize
import matplotlib
import matplotlib.pyplot as plt
```

```python
##HR graph
HR1=[1*x for x in range(0,10000)]  ##0~10초
HR2=[HR1[9999]+2*x for x in range(0,15000)] ##10~25초
HR3=[HR2[14999]-5*x for x in range(0,20000)] ##25~45초
HR4=[HR3[19999]+exp(x/4000) for x in range(0,38000)] ##45~83초
HR5=[HR4[37999]-0.5*x for x in range(0,17000)] ##83~100초
HR6=[HR5[16999]-0.25*x for x in range(0,20000)] ##100~120초
HR=[]
HR.extend(HR1)
HR.extend(HR2)
HR.extend(HR3)
HR.extend(HR4)
HR.extend(HR5)
HR.extend(HR6)
print(len(HR))
HR=np.array(HR)
print(np.shape(HR))
A=max(HR)-min(HR)

HR_1=[x/(2*A)+1 for x in HR]

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(1,1,1)
ax.set_ylim([0.5, 1.5])
xs = np.array([x for x in range(0,120000)])
ys = HR_1


plt.axvline(x=10000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=25000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=45000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=83000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=100000, color='r', linestyle='--', linewidth=0.5)

plt.plot(xs, ys)
plt.title('HR graph')
plt.show()
hr=np.array(HR_1)
print(hr)
```

```python
##HP graph
HR1=[2*x for x in range(0,10000)] ##0~10초
HR2=[HR1[9999]+7*x*abs(math.sin((x/30000)*math.pi)) for x in range(0,15000)] ##10~25초
HR3=[HR2[14999]-(10*x)*abs(math.sin((x/40000)*math.pi)) for x in range(0,20000)] ##25~45초
HR4=[HR3[19999]+2*exp(x/3500) for x in range(0,38000)] ##45~83초
HR5=[HR4[37999]-0.5*x*math.log(x+3) for x in range(0,17000)]##83~100초
HR6=[HR5[16999]-0.25*exp(x/1700) for x in range(0,20000)] ##100~120초
HR=[]
HR.extend(HR1)
HR.extend(HR2)
HR.extend(HR3)
HR.extend(HR4)
HR.extend(HR5)
HR.extend(HR6)
print(len(HR))
HR=np.array(HR)
print(np.shape(HR))
A=max(HR)-min(HR)

HP=[x/(2*A)+1 for x in HR]

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(1,1,1)
ax.set_ylim([0.5, 1.5])
xs = np.array([x for x in range(0,120000)])
ys = HP


plt.axvline(x=10000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=25000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=45000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=83000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=100000, color='r', linestyle='--', linewidth=0.5)

plt.plot(xs, ys)
plt.title('HP graph')
plt.show()
```

```python
##RR graph
HR1=[0.45*x for x in range(0,10000)] ##0~10초
HR2=[HR1[9999]+0.25*x*math.log(x+3) for x in range(0,15000)] ##10~25초
HR3=[HR2[14999]-(2*x)*abs(math.sin((x/40000)*math.pi)) for x in range(0,20000)] ##25~45초
HR4=[HR3[19999]+2*exp(x/4500) for x in range(0,38000)] ##45~83초
HR5=[HR4[37999]-0.05*x*math.log(x+1) for x in range(0,17000)]##83~100초
HR6=[HR5[16999]-0.25*abs(math.sin((x/40000)*math.pi))*x for x in range(0,20000)] ##100~120초
HR=[]
HR.extend(HR1)
HR.extend(HR2)
HR.extend(HR3)
HR.extend(HR4)
HR.extend(HR5)
HR.extend(HR6)
print(len(HR))
HR=np.array(HR)
print(np.shape(HR))
A=max(HR)-min(HR)

RR=[x/(2*A)+1 for x in HR]

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(1,1,1)
ax.set_ylim([0.5, 1.5])
xs = np.array([x for x in range(0,120000)])
ys = RR


plt.axvline(x=10000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=25000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=45000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=83000, color='r', linestyle='--', linewidth=0.5)
plt.axvline(x=100000, color='r', linestyle='--', linewidth=0.5)

plt.plot(xs, ys)
plt.title('RR graph')
plt.show()
```
